```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
filename="Diwali Sales Data.csv"
df=pd.read_csv(filename,encoding="unicode_escape")
data=df
```

```python

data.info()
```

```python
data.shape
```

```python
data.index
```

```python
data.drop(["Status","unnamed1"], axis=1,inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data.dropna(inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data["Amount"]=data["Amount"].astype("int")
data["Amount"].dtype
```

```python
data.info()
```

```python
data.rename(columns={"Marital_Status":"Shaadi"})
```

```python
data.describe()
```

```python
ax=sns.countplot(x="Gender",data=data)
for bar_container in ax.containers:
    ax.bar_label(bar_container)
```

```python
ttt=sns.countplot(x="Age Group", data=data)
for barcon in ttt.containers:
    ttt.bar_label(barcon)
```

```python
test=data.groupby("Gender", as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
test
```

```python
Genamt=sns.barplot(x="Gender",y="Amount",data=test)
for amt in Genamt.containers:
    Genamt.bar_label(amt)
```

```python
agebar=data.sort_values("Age Group", ascending=True)
ageplt=sns.countplot(x="Age Group",data=agebar,hue="Gender")
for age in ageplt.containers:
    ageplt.bar_label(age)
```

SO from the above graph we see that Age group of 26 to 35 made more purchases than the other Age group

```python
age=data.groupby("Age Group", as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
agebar=sns.barplot(x="Age Group",y="Amount",data=age)
for age in agebar.containers:
    agebar.bar_label(age)
```

Women between the Age of 26 and 35 have spent the most on Diwali purchases

States

```python
data=data.sort_values("State")
sns.set(rc={"figure.figsize":(20,5)})
states=sns.countplot(data=data,x="State",hue="Gender")
for st in states.containers:
    states.bar_label(st)
```

```python
states=data.groupby("State",as_index=False)["Amount"].sum().sort_values("Amount",ascending=False).head(10)
states=sns.barplot(x="State",y="Amount",data=states)
for st in states.containers:
    states.bar_label(st)
```

More purchases were from Uttar Pradesh,Maharashtra and Karnataka

```python
states=data.groupby("State",as_index=False)["Orders"].sum().sort_values("Orders",ascending=False).head(10)
states=sns.barplot(x="State",y="Orders",data=states)
for st in states.containers:
    states.bar_label(st)
```

```python
data.columns
```

```python
sns.set(rc={"figure.figsize":(35,5)})
sns.countplot(x="Product_Category",data=data)
```

```python
pd=data.groupby("Product_Category",as_index=False)["Orders"].sum().sort_values("Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
pdplt=sns.barplot(x="Product_Category",y="Orders",data=pd)
for p in pdplt.containers:
    pdplt.bar_label(p)
```

```python
pd=data.groupby("Product_Category",as_index=False)["Amount"].sum().sort_values("Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
pdplt=sns.barplot(x="Product_Category",y="Amount",data=pd)
for p in pdplt.containers:
    pdplt.bar_label(p)
```

Most Purchases were made in Food, Clothing and Electronics.

Summary:

Married Women between the age of 26 and 35 who are from Uttar Pradesh, Maharastra and Karnataka, and work in IT health care and Aviations have made the most purchases on Food,clothing and Electronics During Diwali
